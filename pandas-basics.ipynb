```python
df['DidPlay'] = df.Result =='Play'
df.head(3)
```

```python
df = df.drop(columns=['Result'])
df.tail(2)
```

```python
df.Outlook.value_counts()
```

```python
df[['is_overcast', 'is_rain', 'is_sunny']] = pd.get_dummies(df.Outlook)
df.head(3)
```

```python
df = df.drop(columns=['Outlook'])
df.head(2)
```

```python
df.info()
```

```python
df = df.rename(columns={'Temperature': 'Farenheit'})
df.head()
```

```python
df['Celcius'] = (df.Farenheit - 32) * (5/9)
df.head(3)
```

```python
df['Date'] = pd.to_datetime(df.Date)
df['DateTime'] = pd.to_datetime(df.DateTime)
df.head()
```
